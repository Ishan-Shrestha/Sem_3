# Visualizing data

```python
#Importing external libraries
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
```

### Line Chart

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

# create a line chart, years on x-axis, gdp on y-axis
plt.plot(years, gdp, color='blue', marker='o', linestyle='solid')

# add a title
plt.title("Nominal GDP")

# add a label to the x-axis
plt.xlabel("Years")

# add a label to the y-axis
plt.ylabel("Billions of $")
plt.show()
```

Using concept of matplot in line graph.

Population in Hiroshima after nuclear bombing.

```python
no_of_people = [100, 80, 40, 15, 1]
time = ['8:40 AM','9:40 AM','10:40 AM','11:40 AM','12:40 AM']
plt.plot(time, no_of_people, 'bo-')

plt.xlabel('Time--->')

plt.ylabel('No of people in thousands--->')
plt.show()
```

Using seaborn

```python
flight = sns.load_dataset('flights') # load flight dataset
print(flight)

flight_new = flight.groupby('year').sum(numeric_only=True).reset_index()
# line chart
sns.lineplot(x='year',y='passengers',data=flight_new)
```

## Loading Datasets

```python
# Load dataset

# Load dataset
def load_dataset(file_path):
  return pd.read_csv(file_path)


df = load_dataset("titanic.csv")
df.head(10)
```

### Bar chart

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]

# plot bars with left x-coordinates [0, 1, 2, 3, 4], heights [num_oscars]
plt.bar(range(len(movies)), num_oscars)

plt.title("My Favorite Movies")     # add a title
plt.ylabel("# of Academy Awards")   # label the y-axis

# label x-axis with movie names at bar centers
plt.xticks(range(len(movies)), movies)

plt.show()
```

#### Univariate

```python
sns.countplot(data=df, x='Sex')
df['Survived'].value_counts()
```

#### Bivariate

```python
# Bivariate Analysis
sns.barplot(x='Pclass',y='Age',hue='Sex',data=df)
```

### Pie Chart

```python
 # pie chart

df['Sex'].value_counts().plot(kind='pie',autopct='%0.2f')
```

### Box plot

```python
# Boxplot
# 5 number summary
# median , percentile, min=Q1-1.5*IQR, max=Q3+1.5*IQR, IQR = (Q1-Q3), outlier

sns.boxplot(df['Age'])
```

```python
# Bivarient Analysis


#Box plot (Numerical - Categorical)

sns.boxplot(x='Sex',y='Age',hue='Survived',data=df)
```

### Histogram

```python
# Numberical Value

# Histogram treat as categorical data distribution of data

plt.hist(df['Age'])
```

```python
# Distplot
# probability density function --> The line is kde
# skweness
sns.displot(df['Age'], kind="hist", kde=True)
```

```python
# Bivariant analysis
sns.kdeplot(df[df['Survived'] == 0]['Age'], label='Died')
sns.kdeplot(df[df['Survived'] == 1]['Age'], label='Survived')

plt.title('Age Distribution by Survival')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
plt.show()
```
